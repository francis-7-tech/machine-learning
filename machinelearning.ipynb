**Step 3: Clean and Preprocess the Data**

```python
print("Missing values:")
print(data.isnull().sum())
```

```python
data.drop_duplicates()
```

```python
# Convert 'date' to datetime
data['date'] = pd.to_datetime(data['date'])

# Extract year and month
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month

# Drop the original date column
data = data.drop('date', axis=1)
```

```python
categorical_cols = ['waterfront', 'view', 'condition', 'grade']

for col in categorical_cols:
    data[col] = data[col].astype('category').cat.codes
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select numeric columns only
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
numeric_cols = numeric_cols.drop('price')  # Exclude target variable

scaler = MinMaxScaler()
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])
```

**Step 6: Apply Machine Learning Analysis**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Select important features (you can add more features if you want)
features = ['sqft_living', 'bedrooms', 'bathrooms', 'floors', 'grade']
X = data[features]

# Target variable
y = data['price']
```

```python
# Split the dataset: 70% training, 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create the model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test data
y_pred = model.predict(X_test)
```

```python
# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Root Mean Squared Error (RMSE)
rmse = mean_squared_error(y_test, y_pred, squared=False)

# R-squared (R2 Score)
r2 = r2_score(y_test, y_pred)

# Print the results
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2 Score): {r2:.4f}")
```
